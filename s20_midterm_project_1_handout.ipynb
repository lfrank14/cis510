#Step 2

Add a new column called 'length'. Same as with Titanic passenger names.

```
#your code
ntweets = len(tweet_table)
tweet_list = tweet_table['text'].to_list()

tweet_len = []
for i in range(ntweets):
  tweet = tweet_list[i]
  tweet_len.append(len(tweet))

tweet_table['length'] = tweet_len
```

```
tweet_table.head() #lengths of 101, 122, 19, 84, 38 for first 5 rows
```

# Step 3

Add a column for a count of the hashtags in each tweet. Why? Mostly just a guess they might be useful.

```
#your code
nhashtags = []
for i in range(ntweets):
  tweet = tweet_list[i]
  nhashtags.append(tweet.count('#'))

tweet_table['#'] = nhashtags
```

```
tweet_table.head()   #counts of 1, 3, 0, 1, 1 for first 5 rows
```

# Step 4

Add a column for a count of the exclamation points in each tweet. Why? Ditto.

```
#your code
nexcl = []
for i in range(ntweets):
  tweet = tweet_list[i]
  nexcl.append(tweet.count('!'))

tweet_table['!'] = nexcl
```

```
tweet_table.head()  #counts of 0, 0, 0, 3, 0 for first 5 rows
```

# Step 5

Create columns for the 26 chars and then fill them with the correct values.

```
#your loop here
import string

alpha = list(string.ascii_lowercase)
nalpha = len(alpha)

for i in range(nalpha):
  letter = alpha[i]

  letter_count = []
  for i in range(ntweets):
    tweet = tweet_list[i]
    tweet = tweet.lower() # just in case there are upper case letters in the tweets, too many rows to check manually
    letter_count.append(tweet.count(letter))
  
  tweet_table[letter] = letter_count
```
