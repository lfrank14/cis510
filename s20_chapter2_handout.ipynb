* **`for`**. The '`for`' is special to Python. It shows up in purple to tell us Python recognizes it as something special. It is what gives 'for loop' its jargony name. What would be a better name? Probably '`repeat some steps`'. But we are stuck with `for`.

* **`steves_index`**. The next on the line is '`steves_index`'. I think this is one of the hardest things to understand about loops. I can tell you that it is a variable name. And I hope I am demonstrating that what you call it is up to you. I can also tell you that it is kind of a throw-away variable. We use it in the loop and then don't care about it anymore. Why I believe it is confusing is because behind the scenes, not seen by you, Python is messing with it. Here is the motivation. When I am playing music, I have to remember where I am on the repeats. If it is the first time through, then I repeat. If it is the second time through, I don't repeat. Even I can generally remember whether first or second. But what if I had to repeat 3 times? I could probably remember. What about 10 times? Getting tricky. Python tries to be helpful by giving you a way to see how many times you have repeated. It shows you in steves_index. Check this out.

```python
for steves_index in range(3):
  print(f"I have repeated this many times: {steves_index+1}")  #Notice my use of an f string
```

Kind of confusing, right? It looks like Python is doing `steves_index = 0` to start then `steves_index += 1` thereafter. All hidden from view. Let's use that way of thinking of it for now.

* **`in`**. Next on the line is '`in`'. Annoying and easy to forget. But you have to have it.

* **`range(3)`**. Next on the line is '`range(3)`'. The 'in' goes with this. It says how many times to repeat. Why not simply '`in 3`'? Why do we need '`in range(3)`'? Seems more complicated. It is more complicated :) But I am hoping you can just remember that if you want to repeat something 10 times, you use '`in range(10)`'.

* **`:`**. Finally we have the annoying colon to end the line. You cannot leave it off. And it is easy to forget. Sorry. Just one of those things you have to remember that makes programming infuriating.

##The loop-body

Before discussing what the steps of the loop are (jargony: the loop body), let's use another example. Almost same as one above but I added a new print statement.

```python
for steves_index in range(3):
  print(f"I have repeated this many times: {steves_index}")
print('done')
```

###I am hoping you are surprised

Why wasn't 'done' printed 3 times? Because it is not part of the loop body. How do I know that? Because it was not indented. So the loop body is defined to be the steps (lines) that follow the colon **and are indented**. The line `print('done')` is not indented. Hence, it signals the end of the loop body.

###I'll take a survey at end of course

Do you like this indentation feature or not?

* The case for it: it is simple and easy to read. I can quickly look at the code 
and see what is indented and know this is the loop body.

* The case against it: it is easy to make mistakes. What if I accidently indent the `print('done')` line? It becomes part of the loop. You typically will not get an error immediately. You just will get weird results later.

##Loop pattern: build a new list from an old list

This is another thing that will be common to us this quarter. We have a list. We want a new list that modifies the items in the old list. The old list stays the same.





```python
#set up a test list

test_list = ['a', 'b', 'c', 'd']
```

```python
#store the list length in n

n = len(test_list)
```

```python
new_list = []  #yes, you can just use empty brackets to build an empty list.

for i in range(n):
  item = test_list[i]  #remember Python manipulates i behind the scenes
  new_item = item.upper()
  new_list.append(new_item) #or new_list += [new_item]

print('done')
print(test_list)  #stays the same
print(new_list)   #upper case
```

Here it is as a gist.

```python
test_list = ['a', 'b', 'c', 'd']
n = len(test_list)  #how many elements in test_list

new_list = []  #yes, you can just use empty brackets to build an empty list.
for i in range(n):
  item = test_list[i]  #get value from old list
  new_item = item.upper()  #do something to the item
  new_list.append(new_item)  #add the item to the new list

print('done')
print(test_list)  #stays the same
print(new_list)   #upper case
```

##Wrangling step 2: create a list of name-lengths (use a loop pattern)

So I want to go through each name and get its length in characters (including spaces and punctuation). I want to add the length to a new list.

We have a pattern that creates a new list from an old list so can use that as a start and modify to fit the new problem.

```python
#test_list becomes name_list

n = len(name_list)

name_lengths = []  #new list I want to build

for i in range(n):
  name = name_list[i]
  length = len(name)  #had to add this to template to fit the problem
  name_lengths.append(length)
```

```python
name_lengths[:5]  #you can check against the names by counting characters
```

###Now super easy to add a new column

```python
trimmed_table['Length'] = name_lengths  #adds a new column and sets its values - nice
```

Even though it is a one-liner, the details are a bit tricky. You are not creating a copy of the table. You are doing a memory store on the table itself. Hence it appears on the left hand side of the =.
