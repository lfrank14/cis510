#3. First insight: recast as a Geometry problem
<img src='https://www.dropbox.com/s/9fcc1crlxp19ijt/major_section.png?raw=1' width='300'>

Our goal is to compute how "similar" the 2 lists of numbers are.
Let's take the view that each list of numbers is actually a point in 27-dimension space. Sounds kind of scary already! Let me give you the intuition pretending we only have 2 numbers in our list. Call the two lists A and B. We can view a list as a point on a 2D plot as shown below. Ignore the angle theta for now.

<img src='https://www.dropbox.com/s/7rtuzw37hgl1oi8/ed_vs_cos.png?raw=1' height=300>

We can use the distance d as a measure of their similarity. The smaller the distance, the closer they are. If the distance is 0, they are the same point.

###Good news!
<img src='https://www.sapaviva.com/wp-content/uploads/2017/06/6S.-Euclid-of-Alexandria-ca.320-275-BC-225x225.jpg' height='100'>

Euclid (circa 300BC) figured it out for us. He came up with a formula for computing d. It defines the "Euclidean distance" between 2 points (here called p and q instead of A and B) as follows.

  <p>
<img src='https://www.dropbox.com/s/9wao0kf3u32i3e9/euclidian.png?raw=1'>

Linking our plot above into this:

* n = 2. We have 2 values in each list.
* q1 is the same as x1, p1 is the same as x2.
* q2 is the same as y1, p2 is the same as y2.

The capital (Greek) letter sigma represents summation. I'll write the code out long hand for you.

BTW: this is a useful site that translates mathematical notation into Python: https://amitness.com/2019/08/math-for-programmers/.


```
p = [1,3]  #just guessing from the plot
q = [3,2]  #ditto
n = 2
```

```
greek_sigma = (p[0]-q[0])**2 + (p[1]-q[1])**2  #Python uses ** for exponentiation. In this case, squaring a quantity.
greek_sigma
```

```
d = greek_sigma**.5  #dredging up some old algebra to take the square root using exponentation
d
```

###Does 2.23606797749979 mean the 2 points are close?

Euclidean distance just gives you the distance. Makes no value judgement. We do know that a distance of 0 says the points are exactly the same. So that is the min value. But there is no limit on the upper value.

We will get to a point where we need to reshape a 1D list like old_list into a 2D list (i.e., a nested list) with 1 column and 3 rows. This is another form of data wrangling: getting our data in right shape to feed to a machine-learning algorithm.

<pre>

</pre>

<pre>

Let's see if we can do it using new list from old list pattern.

```
# matrix = list of lists (python does not recognize / have variable class for matrices)
# old_list = [1,2,3]
n = len(old_list)

new_list = []
for i in range(n):
  item = old_list[i]
  new_item = [item]
  new_list.append(new_item)

print(old_list)
print(new_list)
```
