##A way big new idea: variables

With most modern calculators, you can store your arithmetic result in a special holder and then get the value when you need it later. On my calculator, this is shown as an M key standing for memory. Here is online calculator if you want to play with it: http://www.calculatoria.com/?lay=0#c

<img src='https://www.dropbox.com/s/a54oe3i1nm85f4q/Screenshot%202019-09-20%2008.41.14.png?raw=1' height='500'>

The M keys refer to a single memory cell. Python has a special name for memory: *variables*. One big difference is that my calculator has only a single memory cell and it is named M. In python, you can have as many as you like and you get to name them!

Here is the confusing and controversial part. To create a variable in Python you use the equals sign. Why is it confusing? Because the equal sign goes with mathematics and denotes equivalence. Not so in Python. It is controversial for the same reason: it is easy to confuse Python equals with mathematical equals. We will get back to it later.

Let's go through a few steps with the online calculator and see the equivalent in Python using variables.

##How to manipulate strings

The interesting part is manipulating strings once we have them.

Strings have a number of helpful "methods," which are little bits of code you can put after the variable name to return **a copy** of the string's text with some changes applied (the original string stays the same). For example, adding `.upper()` gives a copy with all letters as upper-case:

```python
upit = message.upper()  # change every character in the string to upper case
upit
```

```python
message  #no change
```

##A little glib

That was a big step we just took. I introduced the idea of a method for manipulating data. And the format is:
<pre>
the_data.the_method()
</pre>
Or in our case:
<pre>
message.upper()
</pre>
So the data followed by a period, followed by the method name, followed by a left and right parentheses. Remember that when I use `message` above, I am really doing a calculator MR operation - pull out the data in the memory location.



##Methods can take arguments

Let me show you first then explain.

```python
print(message.count('i') )
message #the argument is the string i
```

The jargony term for what appears between the parens is *arguments*. We have one argument above, the string i. What the `count` method does is count how many times the argument appears in the string. You get an integer back, i.e., the count.

It is possible to have more than one argumet. Let me show you another example.

```python
message.replace('i', 'eye')  #2 arguments
```

So, replace every occurence of the first argument (a string) with the second argument (a string).

##New concept: default arguments

###Second (alternative) way

Python gives you a short-cut means of doing the `+=` thing. It gives you the list `.append()` method. I'll repeat steps above using `append` instead.

```python
new_list = []  #starts empty
```

```python
new_list.append('a')
new_list
```

First let me say either is fine. Use either `+=` or `append`. I tend to prefer `append` because I think it is a bit more explicit about what is going on - you are appending something onto end of list. And you can just say what you want to add without making it a list. I can tell you one downside of `append` that can trip you up. I'll show you.

```python
new_list = []  #starts empty - all good
```

Here I am thinking about new_list += ['a'] but decide to use append instead. It looks cool.

```python
new_list = new_list.append('a') #easy to make mistake but no error message yet
print(new_list)  #Hint that we have troble in River City - None printed and expecting to see list
```

```python
new_list = new_list.append('b') #this gives an error
print(new_list)
```

To me, this error message is mostly non-sensical unless you are steeped in Python jargon. What the heck is 'NoneType'? What is an 'object'? What is an 'attribute'? See what I mean.

Let's click through to Stack Overflow.



So the problem is that the `append` method does its work behind the scenes and returns the value `None`, which is virtually useless in Python as a value to mess with. Looking at this in more detail:
<pre>
new_list = new_list.append('a')
</pre>
Here is what happens:
1.   Python calls the `append` method and adds 'a' at end of new_list. That's all cool.
2.   The `append` method returns `None`. Still ok. `new_list` is `['a']` at this point and that is what we want.
3.   We assign `new_list` to `None`. Not ok! We have just wiped out `new_list` and made it into `None`, definitely not a list.
This will bite us in following code:
<pre>
new_list = new_list.append('b')
</pre>
What we are doing is really `None.append('b')`. Python chokes. You cannot append an item to `None`. It is not a list!

All that said, I still prefer `append` but `+=` is totally ok as well.
